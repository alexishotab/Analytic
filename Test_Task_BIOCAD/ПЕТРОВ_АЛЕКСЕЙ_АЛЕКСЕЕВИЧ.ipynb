#### 4. Выведите таблицу, которая позволит просмотреть первую дату поставки и общее количество поставленной вакцины в рамках Марки вакцины и страны.

```python
data2=data.sort_values(by=['Дата поставки/новости'], ascending=False).groupby(['Марка вакцины','Страна']).agg({'Вакцина (кол-во)': ['sum'],
                                                                                                              'Дата поставки/новости': ['min']})
data2.columns=['Кол-во поставленных вакцин', 'Дата первой поставки']
data2
```

#### 6. Достать домены источников 1, 2, 3
Для столбцов Источников создать столбцы Доменов, где будет извлечен только домен (источник: https://sputnik-abkhazia.ru/Abkhazia/20200320/..., домен: sputnik-abkhazia.ru). В исходном датасете 3 столбца источника и соответственно должно быть создана 3 столбца с доменами.

```python
import numpy as np
data= pd.read_excel('data.xlsx',sheet_name='Data')
data["Домен 1"] = np.nan
new_names=[]
for i in range(len(data.columns)):
    name=data.columns[i].replace('\n','')
    new_names.append(name)
data.columns=new_names

for i in range(len(data['Источник 1'])):
    dom = data.loc[i,'Источник 1'] 
    if type(dom)==str:
        data.loc[i,'Домен 1']  = dom.split('/')[2]
    else: data.loc[i,'Домен 1']  = np.nan

for i in range(len(data['Источник 2'])):
    dom = data.loc[i,'Источник 2'] 
    if type(dom)==str:
        data.loc[i,'Домен 2']  = dom.split('/')[2]
    else: data.loc[i,'Домен 2']  = np.nan

for i in range(len(data['Источник 3'])):
    dom = data.loc[i,'Источник 3'] 
    if type(dom)==str:
        data.loc[i,'Домен 3']  = dom.split('/')[2]
    else: data.loc[i,'Домен 3']  = np.nan
data
```
