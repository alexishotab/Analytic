```python
data= pd.read_csv('dota_hero_stats.csv')
df=data.filter(items=['id','name','roles'])
df
```

```python
l=[]
b=[]
for i in df.roles:
    a=[]
    cntz=0
    for j in range(len(i)):
        if i[j]=="'" and cntz%2==0:
            cntz+=1
            start=j+1
        elif i[j]=="'" and cntz%2==1:
            cntz+=1
            end=j
            a.append(i[start:end])
    for slovo in a:
        if slovo not in l:
            l.append(slovo)
            b.append(1)
        else:
            f=l.index(slovo)
            b[f]+=1   
sns.lineplot(x=l,y=b,)
```

```python
my_stat = pd.read_csv('my_stat.csv')
subset_1= my_stat.iloc[:10,[1,3]]
indexes=[i for i in my_stat.index if i!=5 and i!=1]
subset_2= my_stat.iloc[indexes,[1,3]]
```

```python
subset_1 =my_stat[my_stat.V1>0]
subset_1
```

```python
import numpy as np
my_stat['V5']=my_stat['V1']+my_stat['V4']
my_stat['V6']=np.log(my_stat['V2'])
```

```python
my_stat=my_stat.rename(columns={"V1":"session_value",
               'V2' :'group',
               'V3' :'time',
               'V4' :'n_users'})

my_stat
```

```python
my_stat.iloc[:,0].fillna(0)
my_stat.loc[my_stat.n_users<0,'n_users']=np.nan
med=my_stat.loc[:,'n_users'].median()
my_stat.iloc[:,3].fillna(med)
```

```python
mean_session_value_data=my_stat.groupby('group',as_index=False).agg({'session_value':'mean'}).rename(columns={'session_value':'mean_session_value'})
mean_session_value_data
```
