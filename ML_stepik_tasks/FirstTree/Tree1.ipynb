```python
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.DataFrame({'X_1': [1, 1, 1, 0, 0, 0, 0, 1], 'X_2': [0, 0, 0, 1, 0, 0, 0, 1], 'Y': [1, 1, 1, 1, 0, 0, 0, 0]})
```

```python
clf=tree.DecisionTreeClassifier(criterion = 'entropy')
```

```python
X=data[['X_1','X_2']]
y=data.Y
```

```python
clf.fit(X,y)
```

C помощью команды фит мы обучили наше дерево предсказанию. Икс большое - наши предикторы, игрик малое - зависимая переменная

А сейчас мы визуализируем наше дерево

```python
from sklearn.datasets import load_iris
tree.plot_tree(clf,class_names=["Negative","Positive"])
```
